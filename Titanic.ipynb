### Data Cleaning

Before analyzing the dataset, I checked for missing values and handled them carefully.

- The Age column had 177 missing values, so I filled them using the median since it is more stable for numerical data.
- The Cabin column had too many missing values (687), so it was dropped because it would not provide useful information.
- The Embarked column had only 2 missing values, so I filled them using the mode (most frequent value).

This cleaning step ensures the dataset is complete and ready for analysis.

```python
import pandas as pd

# Fill missing Age with median
df["Age"] = df["Age"].fillna(df["Age"].median())

# Fill missing Embarked with mode (safe version)
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode().iloc[0])

# Drop Cabin safely
if "Cabin" in df.columns:
    df = df.drop("Cabin", axis=1)

# Remove duplicates
df = df.drop_duplicates()

# Check missing values
df.isnull().sum()
```

### Data Analysis

I used groupby() to explore how different factors affected survival.

```python
df.groupby("Sex")["Survived"].mean()
```

```python
df.groupby("Pclass")["Survived"].mean()
```

```python
df.groupby("Pclass")["Age"].mean()
```

```python
import pandas as pd
df["Age_Group"] = pd.cut(
    df["Age"],
    bins=[0, 12, 18, 60, 100],
    labels=["Child", "Teen", "Adult", "Senior"]
)
# Survival rate by age group
df.groupby("Age_Group", observed=False)["Survived"].mean()
```

### Filtering Data

I filtered specific groups to better understand survival patterns.

```python
df[(df["Sex"] == "female") & (df["Survived"] == 1)]
```

```python
df[(df["Age"] < 12) & (df["Survived"] == 1)]
```

```python
df[(df["Pclass"] == 1) & (df["Survived"] == 1)]
```

### Key Insights

- From the analysis, it is clear that females had a much higher survival rate compared to males.

- Passenger class also had a strong impact. People in 1st class were more likely to survive than those in lower classes.

- Children seem to have a relatively higher survival rate, suggesting they may have been given priority.

- The highest survival rate was observed among female passengers in 1st class, especially those who were younger.
